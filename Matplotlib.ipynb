```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.title("Graph in 2D")
plt.scatter(x,y,c="g")

```

```python
#plt plot
a = np.arange(0,50)
b = a*a
plt.plot(a,b,'r')
```

```python
plt.plot(a,b,'b',linestyle='dashed',linewidth = 1, markersize = 12)
```

```python
c=2*a
plt.subplot(2,2,1)
plt.plot(a,c)
plt.subplot(2,2,1)#same subplot location
plt.plot(a,b)
plt.subplot(2,2,3)
plt.plot(a,b)
```

```python
# Trigonometric diagram
import math
k = np.arange(0,5*np.pi,.001)
l = np.sin(k)
n = np.cos(k)
plt.plot(k,l)
plt.plot(k,n)
plt.show()
```

```python
k = np.arange(0,5*np.pi,.001)
l = np.sin(k)
n = np.cos(k)
p = np.linspace(-2*np.pi, 2*np.pi,1000)
m = np.tan(p)
plt.subplot(2,2,1)
plt.plot(k,l)
plt.subplot(2,2,2)
plt.plot(k,n)
plt.bar(k,n)

plt.subplot(2,2,3)
plt.plot(p,m)
```
