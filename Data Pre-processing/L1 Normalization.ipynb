```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
from sklearn.preprocessing import normalize

features = df.drop(columns = (["Outcome"]))
lebel = df["Outcome"]

normalized_data = normalize(features, norm='l1', axis=1)  # Same for L2 Normalization. Just set norm = "L2"
normalized_df = pd.DataFrame(normalized_data, columns=features.columns)
normalized_df["Outcome"] = lebel

print("Original DataFrame:")
print(df)
print("\nL1 Normalized DataFrame:")
print(normalized_df)
```
